# Reflection-Based Content Creation using AutoGen and Groq LLM

## Objective
This notebook demonstrates a reflection-based agentic AI system using AutoGen.
Two agents collaborate to iteratively improve content quality:

1. Content Creator Agent – drafts and revises content on Generative AI.
2. Content Critic Agent – evaluates language clarity and technical correctness.

The agents follow a reflection-based interaction pattern and produce a final
refined Markdown document.



```python
# “This AutoGen-based system demonstrates a reflection-driven multi-agent workflow.
# A Content Creator agent drafts content, while a Content Critic agent evaluates clarity and technical accuracy.
# The agents iteratively refine the content using feedback loops, with turn limits to ensure controlled convergence.
# Groq LLM is used via an OpenAI-compatible endpoint, eliminating dependency on OpenAI APIs.”
```

```python
from autogen import AssistantAgent, UserProxyAgent, GroupChat, GroupChatManager


llm_config = {
    "model": "llama-3.1-8b-instant",
    "api_key": os.environ["OPENAI_API_KEY"],
    "base_url": os.environ["OPENAI_API_BASE"],
    "temperature": 0.3
}
```

```python
#Define Content Creator Agent
content_creator = AssistantAgent(
    name="ContentCreator",
    llm_config=llm_config,
    system_message="""
You are the Content Creator Agent.
Your role is to draft and revise content on Generative AI.
You must:
- Use clear and professional language.
- Ensure technical accuracy.
- Improve content based on critic feedback.
- Produce a final refined Markdown document.
"""
)
```

```python
#Define Content Critic Agent (Reflection-Based)
content_critic = AssistantAgent(
    name="ContentCritic",
    llm_config=llm_config,
    system_message="""
You are the Content Critic Agent.
Your role is to evaluate drafted content.
You should:
- Reflect internally on the content.
- Evaluate language clarity and structure.
- Check technical correctness of Generative AI concepts.
- Provide constructive feedback only.
- Do NOT rewrite the content.
"""
)
```

```python
#Define User Proxy Agent
user_proxy = UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=5
)
```

```python
#Configure Group Chat & Manager
group_chat = GroupChat(
    agents=[user_proxy, content_creator, content_critic],
    messages=[],
    max_round=5
)

manager = GroupChatManager(
    groupchat=group_chat,
    llm_config=llm_config
)
```

```python
#Step 9: Initiate Reflection-Based Conversation
user_proxy.initiate_chat(
    manager,
    message="""
Simulate a reflection-based two-agent workflow.

Topic: Generative AI

Conversation flow:
1. ContentCreator drafts initial content.
2. ContentCritic evaluates and provides feedback.
3. ContentCreator revises based on feedback.
4. ContentCritic evaluates again.
5. ContentCreator finalizes the content.

Constraints:
- Maximum 3 to 5 turns
- Reflection-based feedback
- Final output must be in Markdown
"""
)
```

```python
user_query = "Explain Generative AI with examples."

draft = content_creator.generate_reply(
    messages=[{"role": "user", "content": user_query}]
)

feedback = content_critic.generate_reply(
    messages=[{"role": "user", "content": f"Review this:\n{draft}"}]
)

final = content_creator.generate_reply(
    messages=[{"role": "user", "content": f"Improve using feedback:\n{feedback}\n\n{draft}"}]
)

print(final)
```
